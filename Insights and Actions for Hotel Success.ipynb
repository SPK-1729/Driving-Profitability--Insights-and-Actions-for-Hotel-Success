# Insight 1
#### From the above table, we can identify customers with high credit limits and proactively reach out to them to inquire about their plans for settling their balance.

```python
# Convert 'creditLimit' to numeric explicitly
ds1_cleaned['creditLimit'] = pd.to_numeric(ds1_cleaned['creditLimit'], errors='coerce')
# Convert DataFrame to a DataFrame object
df_customers = pd.DataFrame(ds1_cleaned)
# Plotting
plt.figure(figsize=(10, 6))
df_customers.groupby('country')['creditLimit'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Credit Limits by Country')
plt.xlabel('Country')
plt.ylabel('Total Credit Limit')
plt.show()
```

# Insight 2

#### The bar chart reveals that the majority of customers are based in Australia, the UK, and the USA, each having high credit limits. This indicates strong financial engagement in these regions. Expanding business operations to other countries could potentially increase market reach and drive further profits.


```python
# Convert 'amount' to numeric explicitly
ds6['amount'] = pd.to_numeric(ds6['amount'], errors='coerce')
df_payments = pd.DataFrame(ds6)
# Plotting
plt.figure(figsize=(10, 6))
df_payments.set_index('paymentDate')['amount'].plot(kind='line', marker='o', linestyle='-')
plt.title('Total Payments Over Time')
plt.xlabel('Date')
plt.ylabel('Payment Amount')
plt.grid(True)
plt.show()
```

## Insight 3

#### The line chart reveals the payment trends for the hotel from 2003 to 2005. It shows that:

- **2003**: Payments were at a consistent level except for a dip in the 10th month.
- **2004**: Payments increased significantly in the 1st month.
- **2005**: Payments peaked in the 1st, 4th, and 7th months.

#### This trend suggests that the hotel may need to investigate the factors influencing these monthly fluctuations by reviewing customer comments and feedback. Additionally, understanding the holiday schedule could help in boosting sales during slower months. By addressing these factors, the hotel can optimize revenue collection and enhance financial strategies more effectively.

```python

# Identifying managers
managers = ds2[ds2['employeeNumber'].isin(ds2['reportsTo'].dropna())]
employees = ds2[~ds2['employeeNumber'].isin(managers['employeeNumber'])]

# Data for the pie chart
pie_data = pd.concat([
    pd.DataFrame({'role': 'Manager', 'count': len(managers)}, index=[0]),
    pd.DataFrame({'role': 'Employee', 'count': len(employees)}, index=[1])
])

# Plotting
plt.figure(figsize=(10, 6))
plt.pie(pie_data['count'], labels=pie_data['role'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Managers and Employees')
plt.show()
```

#### From the piechart we can say that only 26% of the job holders are able to control the income of the organization.

```python
customer_count=ds1['customerName'].count()
order_count=ds5['orderNumber'].count()
check_count=ds6['checkNumber'].count()
product_count=ds8['productCode'].count()
amount_collected=ds6['amount'].sum()


results = pd.DataFrame({'Count': [customer_count, order_count, check_count, product_count], 
                       'Amount': [0, 0, amount_collected, 0]}, 
                       index=['Customers', 'Orders', 'Checks', 'Products'])
print(results)
```

```python
# Merge datasets
merged_data = pd.merge(ds5, ds1, on='customerNumber', how='inner')
merged_data = pd.merge(merged_data, ds4, on='orderNumber', how='inner')
merged_data = pd.merge(merged_data, ds2, left_on='salesRepEmployeeNumber', right_on='employeeNumber', how='inner')
merged_data = pd.merge(merged_data, ds3, on='officeCode', how='inner')
merged_data = pd.merge(merged_data, ds6, on='customerNumber', how='inner')
merged_data = pd.merge(merged_data, ds8, on='productCode', how='inner')
merged_data = pd.merge(merged_data, ds7, on='productLine', how='inner')


# Create a pivot table
pivot_table = pd.pivot_table(merged_data,values='amount',index=['customerName','orderNumber','productName','checkNumber','quantityOrdered'],columns=['status'],aggfunc='sum',fill_value=0)




# Display the pivot table
print(pivot_table.head())
print(pivot_table.tail())
```

# Insight 7
The pivot table provides valuable insights into the status of each product order. It allows us to:

- Easily identify which products are on hold, disputed, in process, or cancelled.
- See which customer placed each order.
- Determine the specific products ordered.
- View the associated check number for each order.

This detailed breakdown helps in tracking order statuses, managing customer interactions more effectively, and promptly addressing any issues.
